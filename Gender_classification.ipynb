```python
y=dataset[['gender']]
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
y['gender']=LE.fit_transform(y['gender'])
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X[['forehead_width_cm',	'forehead_height_cm']] = st.fit_transform(X[['forehead_width_cm',	'forehead_height_cm']])
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred =lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
```

```python
acc_score = accuracy_score(y_pred,y_test)
fo_score = f1_score(y_pred,y_test)
pre_score = precision_score(y_pred,y_test)
re_score = recall_score(y_pred,y_test)
score = roc_auc_score(y_pred,y_test)
```
